```python
import pandas as pd
data = pd.read_csv('/content/twitter_sentiment_data.csv')
data
```

```python
data.columns = ['Tweet_id', 'Entity', 'sentiment', 'text']
data
```

```python
data.describe(include='all')
```

```python
data['Entity'].unique()
```

```python
data['sentiment'].unique()
```

```python
entity_types = data['Entity'].value_counts()
entity_types
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(20,15))

sns.barplot(x=entity_types.values, y=entity_types.index)
plt.title('No. of tweets related to Entity')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
sns.catplot(x="Entity", hue="sentiment", kind="count", height=10, aspect=3, data=data)
```

```python
sentiment_types = data['sentiment'].value_counts()
sentiment_types
```

```python
plt.figure(figsize=(16,9))
plt.pie(x=sentiment_types.values, labels=sentiment_types.index, autopct='%.1f%%', explode=[0.1,0.2,0,0])
plt.title('Content Type on social media')
plt.show()
```
